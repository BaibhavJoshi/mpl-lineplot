```python
from matplotlib import pyplot as plt
import random
```

```python
ages_x = [i for i in range(25, 36)]

print(ages_x)
```

```python
dev_y = []

for i in range(0, len(ages_x)):

    dev_y.append(round(random.uniform(35000, 80000), 2))

dev_y.sort()

print(dev_y)
```

```python
py_dev_y = []

for i in range(0, len(ages_x)):

    py_dev_y.append(round(random.uniform(45000, 90000), 2))

py_dev_y.sort()

print(py_dev_y)
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.plot(ages_x, dev_y, linewidth=2, label="All Devs")

plt.plot(ages_x, py_dev_y, linewidth=2, label = "Python Devs")

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.legend()

plt.show()
```
